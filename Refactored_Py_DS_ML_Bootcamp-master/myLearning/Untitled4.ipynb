```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.linspace(0,5,11)
```

```python
y=x**2
```

```python
plt.plot(x,y,'r')
plt.xlabel('X label')

plt.ylabel('X label')

plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x**2,y**2)
```

```python
fig= plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('as')
axes.set_ylabel('as')
axes.set_title('as')
```

```python

fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
m=plt.figure()
ax=m.add_axes([0,0,1,1])
ax.plot(x,y,'purple',lw=1,ls='--' )

ax.legend((0.1,0.1))
ax.set_xlim([0,1])
ax.set_ylim([0,20])
```
